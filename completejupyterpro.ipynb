```python
mean_fixed_acidity = new_dataset.fixed_acidity.mean()
mean_volatile_acidity = new_dataset.volatile_acidity.mean()
mean_citric_acid = new_dataset.citric_acid.mean()
mean_residual_sugar = new_dataset.residual_sugar.mean()
mean_chlorides = new_dataset.chlorides.mean()
mean_free_sulfur_dioxide = new_dataset.free_sulfur_dioxide.mean()
mean_total_sulfur_dioxide = new_dataset.total_sulfur_dioxide.mean()
mean_density = new_dataset.density.mean()
mean_pH = new_dataset.pH.mean()
mean_sulphates = new_dataset.sulphates.mean()
mean_alcohol = new_dataset.alcohol.mean()
mean_quality = new_dataset.quality.mean()
```

```python
print(f'fixed_acidity:{mean_fixed_acidity }')
print(f'volatile_acidity:{mean_volatile_acidity}')
print(f'citric_acid:{mean_citric_acid}')
print(f'residual_sugar:{mean_residual_sugar}')
print(f'chlorides:{mean_chlorides }')
print(f'free_sulfur_dioxide:{mean_free_sulfur_dioxide}')
print(f'total_sulfur_dioxide:{mean_total_sulfur_dioxide}')
print(f'density:{mean_density}')
print(f'pH:{mean_pH}')
print(f'sulphates:{mean_sulphates}')
print(f'alcohol:{mean_alcohol}')
print(f'quality:{mean_quality}')
```

```python
dataset['fixed_acidity'] = dataset['fixed_acidity'].fillna(mean_fixed_acidity)
dataset['volatile_acidity'] = dataset['volatile_acidity'].fillna(mean_volatile_acidity)
dataset['citric_acid'] = dataset['citric_acid'].fillna(mean_citric_acid)
dataset['residual_sugar'] = dataset['residual_sugar'].fillna(mean_residual_sugar)
dataset['chlorides'] = dataset['chlorides'].fillna(mean_chlorides)
dataset['free_sulfur_dioxide'] = dataset['free_sulfur_dioxide'].fillna(mean_free_sulfur_dioxide)
dataset['total_sulfur_dioxide'] = dataset['total_sulfur_dioxide'].fillna(mean_total_sulfur_dioxide)
dataset['density'] = dataset['density'].fillna(mean_density)
dataset['pH'] = dataset['pH'].fillna(mean_pH)
dataset['sulphates'] = dataset['sulphates'].fillna(mean_sulphates)
dataset['alcohol'] = dataset['alcohol'].fillna(mean_alcohol)
dataset['quality'] = dataset['quality'].fillna(mean_quality)
```
